Evaluation
We set 400 as the maximum integer considered.

```python
import pandas as pd
import matplotlib.pyplot as plt

naive_partition = pd.read_csv("naive_runtime_data.txt", sep='\t', skiprows=1, header=None, names=['n', 'partition(n)', 'runtime(n)'])

print(naive_partition)
```

```python
euler_partition = pd.read_csv("euler_runtime_data.txt", sep='\t', skiprows=1, header=None, names=['n', 'partition(n)', 'runtime(n)'])

print(euler_partition)
```

```python
plt.figure(figsize=(10, 10))
plt.plot(naive_partition['n'], naive_partition['partition(n)'], marker='o', linestyle='-', color='b', linewidth=0.25, markersize=1)
plt.xlabel('n')
plt.ylabel('partition(n)')
plt.title('Partition of n : Naive')
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.plot(euler_partition['n'], euler_partition['partition(n)'], marker='o', linestyle='-', color='b', linewidth=0.25, markersize=1)
plt.xlabel('n')
plt.ylabel('partition(n)')
plt.title('Partition of n : Euler')
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.plot(naive_partition['n'], naive_partition['runtime(n)'], marker='o', linestyle='-', color='b', linewidth=0.25, markersize=1)
plt.xlabel('n')
plt.ylabel('runtime(n)')
plt.title('Runtime of partition(n) : Naive')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.plot(euler_partition['n'], euler_partition['runtime(n)'], marker='o', linestyle='-', color='b', linewidth=0.25, markersize=1)
plt.xlabel('n')
plt.ylabel('runtime(n)')
plt.title('Runtime of partition(n) : Euler')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.plot(euler_partition['n'], euler_partition['runtime(n)'], marker='o', linestyle='-', color='red', linewidth=0.25, markersize=1, label='Euler')
plt.plot(naive_partition['n'], naive_partition['runtime(n)'], marker='o', linestyle='-', color='blue', linewidth=0.25, markersize=1, label='Naive')
plt.xlabel('n')
plt.ylabel('runtime(n)')
plt.title('Runtime of partition(n)')
plt.legend()
plt.grid(True)
plt.show()
```
